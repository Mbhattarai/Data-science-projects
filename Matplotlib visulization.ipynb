## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matpoltlib inline
```


* ** Creating a figure object called fig using plt.figure() **
* ** Using add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Titile')
             
```


** Creating a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.55,.2,.2])
```

** Now ploting (x,y) on both axes. And call your figure object to show it.**

```python
ax1.plot(x,y)
ax1.set_xlabel('large X')
ax1.set_ylabel('large Y')
ax1.set_title('large XY')
ax2.plot(x,y)
ax2.set_xlabel('small X')
ax2.set_ylabel('small Y')
ax2.set_title('small XY')
fig
```



** Creating the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4]) 
```

** Now using x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

** now plot by adding the figsize() argument in plt.subplots().**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
